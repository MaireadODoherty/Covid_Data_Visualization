```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
Data = pd.read_csv("C:/Files/Covid/Data/CovidStatisticsProfileHPSCIrelandOpenData.csv")
```

```python
Data.tail()
```

```python
Data['Date'] = pd.to_datetime(Data['Date'], format='%Y/%m/%d')
```

```python
#Data['month'] = Data['Date'].apply(lambda x: x.day)
```

```python
fig, ax1 = plt.subplots()
color = 'tab:blue'
ax1.plot(Data.Date,Data.ConfirmedCovidCases)
#ax.plot(Data.Date, Data.ConfirmedCovidDeaths)
ax1.set_xlabel('Month')
ax1.set_ylabel('Covid Cases', color=color)
monthyearFmt = mdates.DateFormatter('%Y %B')
ax1.xaxis.set_major_formatter(monthyearFmt)

ax1.tick_params(axis='y', labelcolor=color)


color = 'tab:orange'
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Covid Deaths', color=color)  # we already handled the x-label with ax1
ax2.plot(Data.Date, Data.ConfirmedCovidDeaths, color=color)
ax2.tick_params(axis='y', labelcolor=color)

## Rotate date labels automatically
fig.autofmt_xdate()
plt.show()
```
